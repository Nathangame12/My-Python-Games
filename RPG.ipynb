```python
#RPG
import random

# Lista de inimigos com atributos
inimigos = {
    "(basico) Bendy": {"vida": 50, "ataque_min": 5, "ataque_max": 10, "ouro": 5, "xp": 5, "descricao": "Bendy, um pequeno demonio de tinta, sempre dançando e sorrindo!"},
    "(resistente) Ink demon": {"vida": 100, "ataque_min": 8, "ataque_max": 18, "ouro": 20, "xp": 10, "descricao": "A segunda forma de Bendy, mais forte e perigoso!"},
    "(Atormentador) Beast Bendy": {"vida": 165, "ataque_min": 15, "ataque_max": 25, "ouro": 35, "xp": 25, "descricao": "A forma final de Bendy, a mais poderosa de todas!"},
    "(Chefao) Nathan": {"vida": 300, "ataque_min": 12, "ataque_max": 50, "ouro": 50, "xp": 45, "descricao": "O criador do jogo, essa batalha sera boa!"},
    "(basico) Bruno": {"vida": 45, "ataque_min": 10, "ataque_max": 45, "ouro": 15, "xp": 15, "descricao": "Sim."},
    "(resistente) Luiz": {"vida": 300, "ataque_min": 1, "ataque_max": 25, "ouro": 20, "xp": 20, "descricao": "Gastou tanto dinheiro em TF2 que virou uma aberração."},
    "(poderoso) Sonic": {"vida": 220, "ataque_min": 19, "ataque_max": 32, "ouro": 25, "xp": 30, "descricao": "Sonic o ouriço, o mais rapido de todos!"},
    "(chefao) CHARA": {"vida": 450, "ataque_min": 25, "ataque_max": 55, "ouro": 100, "xp": 65, "descricao": "=)"},
    "(deus) Steve do Minecraft no Criativo": {"vida": 1000, "ataque_min": 70, "ataque_max": 150, "ouro": 555, "xp": 250, "descricao": "boa sorte..."}
}

# Armas disponíveis
armas = {
    "Espada Curta": {"dano_min": 10, "dano_max": 20, "preco": 50},
    "AK47": {"dano_min": 22, "dano_max": 50, "preco": 350},
    "Carteira de Trabalho": {"dano_min": 50, "dano_max": 85, "preco": 985},
    "Faz o L!!!": {"dano_min": 60, "dano_max": 90, "preco": 1095},
    "TORÇãO TESTICULAR": {"dano_min": 90, "dano_max": 150, "preco": 1666},
    "Carta Coringa": {"dano_min": 0, "dano_max": 200, "preco": 1450},
    "Pau": {"dano_min": 1, "dano_max": 10, "preco": 20},
    "Determinação": {"dano_min": 112, "dano_max": 300, "preco": 3000},
    "BAN HAMER": {"dano_min": 999, "dano_max": 999999, "preco": 10000}
}

# Atributos iniciais do jogador
vida_jogador = 100
pocoes = 3
habilidade_especial = 2
ouro = 0
xp = 0
nivel = 1
xp_para_proximo_nivel = 30
arma_atual = armas["Espada Curta"]

# Mostrar informações dos oponentes
def mostrar_info():
    print("\n📜 **Informações dos Oponentes:**")
    for nome, atributos in inimigos.items():
        print(f"\n🔹 **{nome}**")
        print(f"   ❤️ Vida: {atributos['vida']}")
        print(f"   ⚔️ Ataque: {atributos['ataque_min']} - {atributos['ataque_max']}")
        print(f"   🏆 Recompensa: {atributos['ouro']} ouro | {atributos['xp']} XP")
        print(f"   📝 Descrição: {atributos['descricao']}")

# Loja para comprar itens
def loja():
    global pocoes, ouro, arma_atual
    while True:
        print("\n🛒 **Loja**")
        print(f"💰 Seu ouro: {ouro}")
        print("[1] Comprar Poção (30 ouro)")
        print("[2] Melhorar Arma")
        print("[0] Sair da Loja")

        escolha = input("Escolha uma opção: ")
        if escolha == "1":
            if ouro >= 30:
                pocoes += 1
                ouro -= 30
                print("🧪 Você comprou uma poção!")
            else:
                print("❌ Ouro insuficiente!")
        elif escolha == "2":
            print("\n⚔️ **Armas Disponíveis:**")
            for i, (nome, atributos) in enumerate(armas.items(), 1):
                print(f"[{i}] {nome} - Dano: {atributos['dano_min']} - {atributos['dano_max']} (💰 {atributos['preco']} ouro)")
            escolha_arma = input("Escolha uma arma para comprar (ou 0 para voltar): ")
            if escolha_arma in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
                arma_selecionada = list(armas.keys())[int(escolha_arma) - 1]
                if ouro >= armas[arma_selecionada]["preco"]:
                    arma_atual = armas[arma_selecionada]
                    ouro -= armas[arma_selecionada]["preco"]
                    print(f"⚔️ Você comprou a {arma_selecionada}!")
                else:
                    print("❌ Ouro insuficiente!")
        elif escolha == "0":
            break
        else:
            print("❌ Escolha inválida!")

# Sistema de batalha
def batalha():
    global vida_jogador, ouro, xp, pocoes, habilidade_especial, nivel, xp_para_proximo_nivel

    vida_jogador = 100 + (nivel - 1) * 10  # Vida aumenta com o nível
    habilidade_especial = 1  # Reseta habilidade especial

    # Escolha do inimigo
    while True:
        print("\nEscolha um inimigo:")
        print("[0] Aleatório")
        for i, nome in enumerate(inimigos.keys(), 1):
            print(f"[{i}] {nome}")
        print("[10] Ver informações dos oponentes")

        escolha = input("Escolha um número: ")
        if escolha == "10":
            mostrar_info()
            continue
        elif escolha == "0":
            inimigo_nome, inimigo = random.choice(list(inimigos.items()))
            break
        elif escolha in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
            inimigo_nome = list(inimigos.keys())[int(escolha) - 1]
            inimigo = inimigos[inimigo_nome]
            break
        else:
            print("❌ Escolha inválida! Tente novamente.")

    vida_inimigo = inimigo["vida"]
    print(f"\n⚔️ Você enfrentará o {inimigo_nome}!")

    while vida_jogador > 0 and vida_inimigo > 0:
        print(f"\n❤️ Sua vida: {vida_jogador} | 👹 Vida do {inimigo_nome}: {vida_inimigo}")
        print(f"🧪 Poções: {pocoes} | 🌟 Habilidade Especial: {habilidade_especial}")

        print("\n[1] Atacar ⚔️  [2] Defender 🛡️  [3] Usar Poção 🧪  [4] Habilidade Especial 🌟")
        acao = input("Escolha uma ação: ")

        if acao == "1":
            dano = random.randint(arma_atual["dano_min"], arma_atual["dano_max"])
            vida_inimigo -= dano
            print(f"💥 Você atacou e causou {dano} de dano!")
        elif acao == "2":
            print("🛡️ Você se defendeu! O dano recebido será reduzido pela metade.")
        elif acao == "3" and pocoes > 0:
            vida_jogador += random.randint(15, 30)
            pocoes -= 1
            print("🧪 Você usou uma poção!")
        elif acao == "4" and habilidade_especial > 0:
            vida_inimigo -= random.randint(30, 50)
            habilidade_especial -= 1
            print("🌟 Habilidade especial usada!")

        if vida_inimigo > 0:
            ataque = random.randint(inimigo["ataque_min"], inimigo["ataque_max"])
            vida_jogador -= ataque // (2 if acao == "2" else 1)

    if vida_jogador > 0:
        print(" Você venceu!")
        ouro += inimigo["ouro"]
        xp += inimigo["xp"]
        if xp >= xp_para_proximo_nivel:
            nivel += 1
            xp -= xp_para_proximo_nivel
            xp_para_proximo_nivel += 20
            print(" Você subiu de nível!")

while True:
    batalha()
    if input("\nContinuar jogando? (s/n): ").lower() != "s":
        print("Até a próxima!")
        break
    loja()
```

**Modo freeplay**

```python
#RPG (freeplay)
import random

# Lista de inimigos
inimigos = {
    "Goblin": {"vida": 80, "ataque_min": 5, "ataque_max": 15},
    "Troll": {"vida": 120, "ataque_min": 8, "ataque_max": 18},
    "Dragão": {"vida": 150, "ataque_min": 10, "ataque_max": 25}
    #não tem como treinar com o melhor de todos- kk
}

def batalha():
    vida_jogador = 100
    pocoes = 2
    habilidade_especial = 1  # Só pode usar uma vez e bla bla

    # Escolha do inimigo
    print("\nEscolha um inimigo para treinar:")
    for i, nome in enumerate(inimigos.keys(), 1):
        print(f"[{i}] {nome}")

    escolha = input("[0] Aleatório | Escolha um número: ")
    if escolha == "0":
        inimigo_nome, inimigo = random.choice(list(inimigos.items()))
    elif escolha in ["1", "2", "3"]:
        inimigo_nome = list(inimigos.keys())[int(escolha) - 1]
        inimigo = inimigos[inimigo_nome]
    else:
        print("Escolha inválida, enfrentando um inimigo aleatório!")
        inimigo_nome, inimigo = random.choice(list(inimigos.items()))

    vida_inimigo = inimigo["vida"]
    print(f"\n⚔️ Você enfrentará um {inimigo_nome} com {vida_inimigo} de vida!")

    while vida_jogador > 0 and vida_inimigo > 0:
        print(f"\n Sua vida: {vida_jogador} | Vida do {inimigo_nome}: {vida_inimigo}")
        print(f"Poções: {pocoes} | Habilidade Especial: {habilidade_especial}")

        print("\n[1] Atacar   [2] Defender   [3] Usar Poção   [4] Habilidade Especial ")
        acao = input("Escolha uma ação: ")

        # Turno do jogador
        if acao == "1":
            dano = random.randint(10, 20)
            vida_inimigo -= dano
            print(f" Você atacou e causou {dano} de dano!")
        elif acao == "2":
            print(" Você se defendeu! O dano recebido será reduzido.")
        elif acao == "3":
            if pocoes > 0:
                cura = random.randint(15, 30)
                vida_jogador += cura
                pocoes -= 1
                print(f" Você usou uma poção e recuperou {cura} de vida!")
            else:
                print(" Você não tem mais poções!")
        elif acao == "4":
            if habilidade_especial > 0:
                dano_especial = random.randint(30, 50)
                vida_inimigo -= dano_especial
                habilidade_especial -= 1
                print(f" Você usou sua habilidade especial e causou {dano_especial} de dano!")
            else:
                print(" Você já usou sua habilidade especial!")
        else:
            print(" Escolha inválida! Você perdeu a vez.")

        # Turno do inimigo
        if vida_inimigo > 0:
            ataque_inimigo = random.randint(inimigo["ataque_min"], inimigo["ataque_max"])
            if acao == "2":
                ataque_inimigo //= 2
            vida_jogador -= ataque_inimigo
            print(f" O {inimigo_nome} atacou e causou {ataque_inimigo} de dano!")

    # Resultado da batalha (muito facil de ganhar isso;-;)
    if vida_jogador > 0:
        print(f"\n Você derrotou o {inimigo_nome}! Vitória! ")
    else:
        print(f"\n O {inimigo_nome} venceu... Você foi derrotado! ")

while True:
    batalha()
    continuar = input("\n Deseja lutar novamente? (s/n): ").lower()
    if continuar != "s":
        print("\n Obrigado por jogar! Até a próxima!")
        break
```
