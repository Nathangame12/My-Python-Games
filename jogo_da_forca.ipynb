```python
#Jogo da forca
import random

#Lista de palavras para o jogo
palavras = ["python", "programacao", "desenvolvedor", "computador", "teclado", "pneumoultramicroscopicossilicovulcaniotico", "sim", "portugues", "borracha", "barbeiro", "transmutacao"]

#Escolhe uma palavra aleatória
palavra_secreta = random.choice(palavras)
letras_descobertas = ["_"] * len(palavra_secreta)
tentativas_restantes = 6
letras_erradas = []

print("Bem-vindo ao Jogo da Forca! você tem 6 tentativas!")
print(" ".join(letras_descobertas))

#Loop do jogo
while tentativas_restantes > 0 and "_" in letras_descobertas:
    letra = input("\nDigite uma letra: ").lower()

    if letra in palavra_secreta:
        for i, l in enumerate(palavra_secreta):
            if l == letra:
                letras_descobertas[i] = letra
    else:
        letras_erradas.append(letra)
        tentativas_restantes -= 1
        print(f"❌ Letra errada! Você tem {tentativas_restantes} tentativas restantes.")

    print("Palavra: " + " ".join(letras_descobertas))
    print("Letras erradas:", " ".join(letras_erradas))

#Resultado do jogo
if "_" not in letras_descobertas:
    print("\n🎉 Parabéns! Você acertou a palavra:", palavra_secreta)
else:
    print("\n💀 Você perdeu! A palavra era:", palavra_secreta)
```
