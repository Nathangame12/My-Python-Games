```python
#jogo de pedra, papel e tesoura
import random

# Opções do jogo
opcoes = ["pedra", "papel", "tesoura"]

# Variáveis de pontuação
pontos_jogador = 0
pontos_computador = 0

print("Bem-vindo ao Pedra, Papel e Tesoura!")
print("Digite 'x' para encerrar o jogo")

while True:
    # Entrada do usuário
    jogador = input("\nEscolha pedra, papel ou tesoura: ").lower()

    # Opção de sair do jogo
    if jogador == "x":
        print("\nJogo encerrado! Placar final:")
        print(f"Você: {pontos_jogador} | Computador: {pontos_computador}")
        break

    # Verifica se a escolha do usuário é válida
    if jogador not in opcoes:
        print("...as opções são PEDRA PEPEL OU TESOURA! não qualquer outra coisa alem disso...como", jogador)
        continue

    # Escolha do computador
    computador = random.choice(opcoes)

    print(f"\nVocê escolheu: {jogador}")
    print(f"Computador escolheu: {computador}")

    # Verifica o vencedor
    if jogador == computador:
        print("Empate!")
    elif (jogador == "pedra" and computador == "tesoura") or \
         (jogador == "papel" and computador == "pedra") or \
         (jogador == "tesoura" and computador == "papel"):
        print("Você ganhou!")
        pontos_jogador += 1
    else:
        print("Você perdeu!")
        pontos_computador += 1

    # Mostra o placar
    print(f"Placar: Você {pontos_jogador} | Computador {pontos_computador}")
```
